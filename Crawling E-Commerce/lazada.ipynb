```python
import pandas as pd
import re

from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.common.by import By

cols = ['Nama', 'Harga', 'Diskon', 'Terjual', 'Banyak Review', 'Kota/Kab', 'Other']
df = pd.DataFrame(columns = cols)

service = Service("C:\Program Files\edgedriver_win64")
driver = webdriver.Edge(service = service)

url = 'https://www.lazada.co.id/jual-home-entertainment/?spm=a2o4j.home.cate_8.9.f6ef53e0A5K2GT'

driver.get(url)
driver.implicitly_wait(5)
```

```python
res = driver.find_elements(By.XPATH, "//div[@class='_17mcb']")
list_items = res[0].text.split('\n')

i = 0
while i < len(list_items):
    nama = ''
    harga = ''
    diskon = ''
    terjual = ''
    banyak_review = ''
    kota_kab = ''
    other = ''

    count = i
    j = 0
    while (j < 7) and (count + j < len(list_items)):

        if j == 0:
            nama = list_items[count + j]

        elif j == 1:
            harga = list_items[count + j]

        elif 'Off' in list_items[count + j]:
            diskon = list_items[count + j].split(' ')[0]

        elif 'sold' in list_items[count + j]:
            terjual = list_items[count + j].split(' ')[0]

        elif bool(re.search(r'\(\d+\)', list_items[count + j])):
            banyak_review = list_items[count + j][1:len(list_items[count + j]) - 1]

        elif any(substring in list_items[count + j] for substring in ['Kab', 'Kab.', 'Kabupaten', 'Kota']):
            kota_kab = list_items[count + j]

        elif any(substring in list_items[count + j] for substring in ['Save', 'save']):
            other = list_items[count + j]

        j += 1

    temp = [nama, harga, diskon, terjual, banyak_review, kota_kab, other]

    i += len([_ for _ in temp if _ != ''])

    df.loc[len(df)] = temp
```
